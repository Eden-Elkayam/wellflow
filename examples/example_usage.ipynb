```python
import wellflow as wf

# Load example data
measurements = wf.read_plate_measurements("example_plate.xlsx")

design = wf.read_plate_design("example_design.xlsx")

# Merge
df = wf.merge_measurements_and_conditions(measurements, design)

# Process
df = wf.add_blank_value(df, window=2)
df = wf.add_smoothed_od(df, window=2)
df = wf.add_growth_rate(df, window=3)
df = wf.add_max_growth_rate(df)

df.head()
```
